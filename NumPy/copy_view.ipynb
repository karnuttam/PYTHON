## Copy() vs View()

```python
import numpy as np
```

```python
var = np.array([45,23,96,35]);
new_var = var.copy();
var[1] = 50;# Not change the copied data
print("Original: ", var);
print("Copied: ", new_var);
```

```python
var_2 = np.array([45, 36,98,14,45]);
new_var_2 = var_2.view();
var_2[1] = 56; # Also changed the viewed data
print("Original: ",var_2);
print("Viewed: ", new_var_2);
```
