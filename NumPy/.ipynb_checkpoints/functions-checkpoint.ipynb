# Functions

## rand()

```python
# rand(): the function is used to generate a random value between 0 to 1.

import numpy as np
var = np.random.rand(4)

print(var)
```

```python
var2 = np.random.rand(2,4)
print(var2);
```

## randn()


```python
# randn(): the function is used to generate a random value close to  zero.This my return positive or negative numbers as well.
var3 = np.random.randn(5);
print(var3); 
```

## ranf()
### ranf(): the function for doing random sampling in numpy. It returns an array specified shape and fills it with random floats in the half open interval [0.0, 1.0)

```python
var4 = np.random.ranf(5);
print(var4);
```

## randint()
### randint(): the function is used to generate a random number between a given number

```python
# np.random.randint(min, max, total_values)
var5 = np.random.randint(4, 30, 10);
print(var5);
```
