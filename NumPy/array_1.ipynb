## Array 

```python
import numpy as np
l =[]
for i in range(1,5):
    int_1 = int(input("Enter: "))
    l.append(int_1)
y = np.array(l);
print(y);
#print(np.array(l));
print(y.ndim);
```

```python
arr2 = np.array([[45,56,89,25],[78,36,62,59]])
print(arr2);
print("The dimension of the given array is",arr2.ndim);
```

```python
arr3 = np.array([[[45,36,96],[56,25,36], [45,89,36]]])
print(arr3);
print("The dimension of the given array is",arr3.ndim);
```

```python
arr4 = np.array([56,23,45,36],ndmin = 10)
print(arr4);
print("The dimension of the given array is", arr4.ndim);
```

# Zeros

```python
arr_zero = np.zeros(4);
print(arr_zero);
```

```python
arr_zero = np.zeros((4,6));
print(arr_zero);
```

```python
arr_ones = np.ones(4);
print(arr_ones);
```

# Empty

```python
arr_emp = np.empty(6);
print(arr_emp);
```

# Range

```python
arr_rn = np.arange(5);
print(arr_rn);
```

# Diagonal

```python
arr_diag = np.eye(4);
print(arr_diag);
```

```python
arr_diag = np.eye(3,5);
print(arr_diag);
```
