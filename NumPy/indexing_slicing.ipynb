## Indexing

```python
import numpy as np
```

```python
var = np.array([45,65,989,25]);
print(var);
print(var[2]);
```

```python
var = np.array([[56,36,52,89],[96,36,69,45]]);
print(var);
print(var[1,3]);
```

```python
var = np.array([[[45,36,89,25],[86,96,65,32]]]);
print(var);
print(var[0,1,2]);
```

## Slicing

```python
var = np.array([78,56,96,36,46]);
print(var);
print(var[1:3]);
```

```python
var = np.array([78,56,96,36,46]);
print(var);
print(var[1:6:2]);
```

```python
var = np.array([[45,63,98,25],[78,36,95,24]]);
print(var);
print(var[1,1:3]);
```

```python
var = np.array([[[56,35,24,15],[47,56,32,96],[15,37,25,36]]])
print(var);
print(var[0,2,1:3]);
```
