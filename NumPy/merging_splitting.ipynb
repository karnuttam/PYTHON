# Concatenation

```python
import numpy as np
```

```python
var_1 = np.array([45, 63,98,65]);
var_2 = np.array([45, 21, 34, 78]);
print(var_1);
print(var_2);
print("The concatenated array");
new = np.concatenate((var_1,var_2));
new_2 = np.stack((var_1,var_2),axis=1);# use stack with axis to convert 2D array with merge
print(new);
print(new_2);
```

```python
var_1 = np.array([[45,26,35],[78,63,95]])
var_2 = np.array([[45,25,36],[15,28,36]])
print(var_1);
print(var_2);
print("The concatenated array");
new_arr_1 = np.concatenate((var_1,var_2), axis = 1);# Along rows
new_arr_2 = np.concatenate((var_1,var_2), axis = 0);# Along columns
print(new_arr_1);
print(new_arr_2);
```

# Stack

```python
var_1 = np.array([45, 63,98,65]);
var_2 = np.array([45, 21, 34, 78]);
print(var_1);
print(var_2);
print("The concatenated array");
new = np.concatenate((var_1,var_2));
new_2 = np.stack((var_1,var_2),axis=1);# use stack with axis to convert 2D array with merge
print(new);
print(new_2);
```

```python
var_1 = np.array([45, 63,98,65]);
var_2 = np.array([45, 21, 34, 78]);
print(var_1);
print(var_2);
print("The concatenated array");
new = np.concatenate((var_1,var_2));
new_2 = np.hstack((var_1,var_2));# horizontally merge
new_3 = np.vstack((var_1,var_2));# vertically merge
new_4 = np.dstack((var_1,var_2));# depth/height along merge
print(new);
print("Horizontal:",new_2);
print("Vertical:", new_3);
print("depth:", new_4);
```

## Split array

```python
var = np.array([45,36,58,25]);
print(var);
arr = np.array_split(var, 3);
print("Splitted array");
print(arr);
print(arr[1]);
print(type(arr));
```

```python
var = np.array([[45,36],[58,25],[58,69]]);
print(var);
arr = np.array_split(var, 3);
print("Splitted array");
print(arr);
print(arr[1]);
print(type(arr));
```

```python
var = np.array([[45,36],[58,25],[58,69]]);
print(var);
arr = np.array_split(var, 3);
arr_2 = np.array_split(var, 3, axis=1);
print("Splitted array");
print(arr);
print(arr[1]);
print(type(arr));

print();
print(arr_2);
```
