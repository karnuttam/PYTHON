## Shape

```python
import numpy as np
var = np.array([45, 63, 98, 25]);
print(var);
print(var.shape);
```

```python
var_2 = np.array([[54,8,6,],[8,6,9]]);
print(var_2);
print(var_2.shape);
```

```python
var_3 = np.array([58,69,32,65], ndmin = 10);
print(var_3);
print(var_3.ndim);
print(var_3.shape);
```

## Reshape

```python
var_4 = np.array([56, 45, 89,46, 3, 2]);
print(var_4);
print(var_4.ndim);
new = var_4.reshape(2, 3);
print(new);
print(new.ndim);
```

```python
var_4 = np.array([56, 45, 89,46, 3, 2,5,8,9, 5, 6, 8]);
print(var_4);
print(var_4.ndim);
# reshape(depth, rows, columns)
#total elements = depth*rows*columns
new = var_4.reshape(2, 3, 2);
print(new);
print(new.ndim);
print(new.shape);

# convert into ond dimension array again
one_d = new.reshape(-1);
print(one_d);
print(one_d.shape);
```

,## Broadcasting
,,### Broadcasting lets NumPy perform operations on arrays of different shapes without copying data and element-wise, efficiently.

```python
var_1 = np.array([89,36,15,96,25]);
var_2 = np.array([5,6,8,9,6]);
new = var_1 + var_2;
print(new);
```

```python
var_1 = np.array([89,36,15,96,25]);
var_2 = np.array([[5],[77],[96],[58],[36]]);
print(var_1);
print(var_2);
new = var_1 + var_2;
print("Addition");
print(new);
print(new.shape);
print(new.ndim);
print("Multiplication");
print(var_1 * var_2);
print("Division");
print(var_1 / var_2);
```
