# Arithmetic Operations

## 1 - Dimensional Array

```python
import numpy as np
var1 = np.array([45,56,14,89]);
new = var1 + 66;
print(var1);
print(new);
```

```python
var1 = np.array([45,65,36,58]);
var2 = np.array([58,69,32,25]);
new = var1 + var2;
print(var1);
print(var2);
print(new);
```

```python
var1 = np.array([45,65,36,58]);
var2 = np.array([58,69,32,25]);
new = np.add(var1, var2);
print(var1);
print(var2);
print(new);
```

```python
var1 = np.array([45,65,36,58]);
var2 = np.array([58,69,32,25]);
new = var1 % 2;

print(var1);
print(var2);
print(new);
```

```python
var1 = np.array([45,65,36,58]);
var2 = np.array([58,69,32,25]);
new = np.add(var1, var2);
multiply = np.multiply(var1, var2);
print(var1);
print(var2);
print(new);
print(multiply);
```

## 2 - Dimensional Array

```python
var = np.array([[54,25,78],[96,36,77]]);
new = var + 6;
print(var);
print(new);
```

```python
var = np.array([[54,25,78],[96,36,77]]);
new = np.reciprocal(var);
print(var);
print(new);
```
