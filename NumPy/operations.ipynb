## Shuffle

```python
import numpy as np
```

```python
var = np.array([1,2,3,4,5]);
np.random.shuffle(var);
print(var)
```

# Unique

```python
var_2 = np.array([1,1,1,2,2,5,8,6,3]);
new = np.unique(var_2, return_index=True, return_counts=True);
print(var_2);
print(new);
```

# Resize

```python
var_3 = np.array([45,65,25,36]);
y = np.resize(var_3, (3,2));
print(y);
```

# flatten

```python
var_3 = np.array([45,65,25,36]);
y = np.resize(var_3, (3,2));
print(y);
print(y.flatten(order="F"));
print(np.ravel(y, order="A"));
```

# Insert

```python
arr = np.array([45,25,15,96,36,78]);
print(arr);
new_arr = np.insert(arr, (2,5), 99);
new_arr_2 = np.append(arr, 6.56);
print(new_arr);
print(new_arr_2);
```

```python
arr_2 = np.array([[45,23,69,11],[56,98,36,75]]);
print(arr_2);
new_arr = np.insert(arr_2, 2, [88,55,86,66],axis=0);
new_arr_2 = np.append(arr_2, [[45,63,89,22]], axis = 0);
print(new_arr);
print(new_arr_2);
```

# Delete

```python
var = np.array([12,15,13,17]);
print(var);
new_var = np.delete(var, 2);
print(new_var);
```
