## Iterations

```python
import numpy as np
```

```python
var = np.array([45,36,78,96,35]);
print(var);
```

```python
var = np.array([12,45,63,78,96]);
print(var);
for i in var:
    print(i);
```

```python
var =  np.array([[45,12,96,36],[45,96,31,75],[47,62,19,38]]);
print(var);
for i in var:
    for k in i:
        print(k);
```

```python
var =  np.array([[[45,12,96,36],[45,96,31,75],[47,62,19,38]]]);
print(var);
for i in var:
    for k in i:
        for j in k:
            print(j);
```

## nditer()

```python
var = np.array([45,25,63,78]);
print(var);
for i in np.nditer(var):
    print(i);
```

```python
var = np.array([[[45,25,63,78],[45,36,95,78]]]);
print(var);
for i in np.nditer(var, flags=['buffered'],op_dtypes = ['S10'], casting="unsafe"):
    print(i);
```
