# Matrix

```python
import numpy as np
```

```python
#var_1 = np.matrix([[1,5,6,2],[5,6,8,6]]);
#var_2 = np.matrix([[4,3,6,7],[8,9,6,7]]);
var_1 = np.matrix([[1,2],[5,7]]);
var_2 = np.matrix([[4,3],[8,9]]);

print(var_1);
print(var_2);
print(type(var_1));
print("Addition");
print(var_1 + var_2);
print("Multiplication");
print(var_1.dot(var_2));
print("Transpose");
print(np.transpose(var_1));
print("Swapaxes");
print(np.swapaxes(var_2, 0, 1));

var_3 = np.matrix([[45,25],[78,14]])
print(var_3);
print("Inverse");
print(np.linalg.inv(var_3));

print("power of matrix");
print(np.linalg.matrix_power(var_3, 0)); # gives identity matrix
print(np.linalg.matrix_power(var_3, 2));
print(np.linalg.matrix_power(var_3, -2));# gives inverse*power
print("Determinant");
print(np.linalg.det(var_3));
```

# Array

```python
var_1 = np.array([[1,5,6,2],[5,6,8,6]]);
var_2 = np.array([[4,3,6,7],[8,9,6,7]]);

print(var_1);
print(var_2);
print(type(var_1));
print("Addition");
print(var_1 + var_2);
print("Multiplication");
print(var_1*var_2);
```
