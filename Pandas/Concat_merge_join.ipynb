## Merging

```python
import numpy as np;
import pandas as pd;
```

```python
employees = {
    "Employee_ID": [1001, 1002,1003,1004, 1005],
    "Name": ["Rohit Kukereja","Raman Singh", "Tom Holland", "Robert Downey Junior", "Christopher Nolan"],
    "Department": ["Science", "Technical", "Film Making", "Direction", "Acting"]
    }
df = pd.DataFrame(employees);
print(df);
```

```python
salary = {
    "Employee_ID": [1001, 1002, 1003, 1006, 1007],
    "Salary":[12345,45678,78945,25896,478956],
    "Bonus": [5624,4587,9658,3665,1478]
    
}
df_2 = pd.DataFrame(salary);
print(df_2);
```

```python
print(pd.merge(df, df_2));
print("\nOn the basis of comman ID");
print(pd.merge(df, df_2,  on="Employee_ID", how = "inner"));
print("\nPrint all data on the basis of ID");
print(pd.merge(df, df_2, on="Employee_ID", how="outer"));
print("\nOn the basis of employees");
print(pd.merge(df,df_2, on="Employee_ID", how="left"));
print("\nOn the basis of salary");
print(pd.merge(df, df_2, on="Employee_ID", how="right"));
```

## Concatenation

```python

employees = {
    "Employee_ID": [1001, 1002,1003,1004, 1005],
    "Name": ["Rohit Kukereja","Raman Singh", "Tom Holland", "Robert Downey Junior", "Christopher Nolan"],
    "Department": ["Science", "Technical", "Film Making", "Direction", "Acting"]
    }
df = pd.DataFrame(employees);
print(df);
```

```python
salary = {
    "Employee_ID": [1001, 1002, 1003, 1006, 1007],
    "Salary":[12345,45678,78945,25896,478956],
    "Bonus": [5624,4587,9658,3665,1478]
    
}
df_2 = pd.DataFrame(salary);
print(df_2);
```

```python
print(pd.concat([df, df_2]));
```

```python
print(pd.concat([df, df_2], axis = 1));
```

## Joining two DataFrames

```python
df1 = pd.DataFrame({
    "Name": ["Tom", "Bob", "Jack", "Robert"]
}, index = [1, 2, 3, 4]);
print(df1);
df2 = pd.DataFrame({
    "Score": [99,98,97,96]
}, index = [2, 3, 4, 5])
print(df2);
```

```python
print(df1.join(df2));
print("\n");
print(df1.join(df2, how="outer"));
print("\n");
print(df1.join(df2, how="inner"));
```
