```python
def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
        if i == ")":
            check = False
            return data
        if check == True:
            data = data + i
        if i == "(":
            check = True
    return data;
```

```python
print(df["Title"].apply(extract_episodes));
```

```python
df["Episodes"] = df["Title"].apply(extract_episodes);
print(df);
```

```python
df["Episodes"] = df["Episodes"].str.replace(" eps","");
```

```python
def extraction_time(txt):
    check = False
    data = "";
    for i in range(len(txt)):
        if txt[i] == ")":
            for j in range(i + 1, i + 20):
                data = data + txt[j];
            return data;
```

```python
print(df["Title"].apply(extraction_time));
```

```python
df["Total Time"] = df["Title"].apply(extraction_time);
```
