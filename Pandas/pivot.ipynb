## PIVOT TABLES

```python
import numpy as np;
import pandas as pd;
```

```python
data = {
    "Date": pd.date_range("2023-01-01", periods=20),
    "Product": ["A", "B", "C", "D"] * 5,
    "Region": ["East", "West", "North", "South", "East", "North", "South", "East", "West",
               "North", "South", "East", "West", "North", "South", "East", "West","North", "South", "North"],
    "Sales": np.random.randint(100, 1000, 20),
    "Units": np.random.randint(10, 100, 20),
    "Rep": ["John", "Marry", "Bob", "ALice", "John", "Marry", "Bob", "Alice", "John", "Marry","Bob", "Alice",
            "Bob", "Alice", "John", "Marry", "Bob", "Alice", "John", "Alice"]
    
}
df = pd.DataFrame(data);
df["Month"] = df["Date"].dt.month_name();
df["Quarter"] = "Q" + df["Date"].dt.quarter.astype(str);
print(df);
```

```python
print(pd.pivot_table(df, values="Sales", index="Region", columns="Product", aggfunc="median"));
```

```python
print(pd.pivot_table(df, values=["Sales","Units"], index="Region", columns="Product"));
```

## Cross Tabs

```python
print(pd.crosstab(df["Region"], df["Product"]));
```
