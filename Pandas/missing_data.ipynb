## Missing Datas

```python
import numpy as np
import pandas as pd
```

```python
data = {
    "A": [12,56, np.nan, 56, 36],
    "B": [47,np.nan, 56,89,23],
    "C": [58,25,36,np.nan,np.nan],
    "D": [23,65,np.nan,56,75]
}
df = pd.DataFrame(data);
print(df);
```

```python
print(df.isna());
```

```python
values = {"A": 0, "B": 2, "C":3, "D":4};
print(df.fillna(value = values));
```

```python
print(df);
```

```python
print(df.fillna(0));
```

```python
print(df.dropna());
```

```python
print(df.dropna(thresh=2)); # atleast null value
```

```python
print(df.fillna(df.mean()));
```
