## DaraFrame

```python
"""
A DataFrame is a two-dimensional, labeled data structure in Python provided by the Pandas 
library.Basically it is a collectioon of Series used in Pandas>
It is similar to a table in a database or an Excel sheet, where:
1.Rows represent records.
2.Columns represent different fields (variables)
3.Both rows and columns have labels (index and column names)
"""
```

```python
import numpy as np
import pandas as pd
```

```python
dict = {
    "Name": ["Supriya", "Prince", "Uttam", "Nisha"],
    "Age": [21,18,22,23],
    "Marks":[99,99,99,99],
    "College":["PK Roy","PW College","IIIT","Delhi University"]
  
}
df = pd.DataFrame(dict);
print(df);
```

```python
data_list=[
    ["Supriya", "Prince", "Uttam", "Nisha"],
    [21,18,22,23],
    [99,99,99,99],
    ["PK Roy","PW College","IIIT","Delhi University"]
]
df_2 = pd.DataFrame(data_list);
print(df_2);
```

```python
data_list=[
    ["Supriya", 21, "Dhanbad", 12542],
    ["Uttam", 22, "Bhagalpur", 4458585],
    ["Nisha", 23, "Delhi", 145555],
    ["Sanyog", 22, "Patna", 44555],
    ["Prince", 18, "Gaya", 78545]
]
df = pd.DataFrame(data_list);
print(df);
```

```python
columns=["Name","Age", "City", "Salary"];
df = pd.DataFrame(data_list, columns=columns);
print(df);
```

```python
print(df["Name"]);
print(df[["Name", "City"]])
```

```python
df["Designation"] = ["Engineer","Engineer", "Doctor", "IAS Officer", "Engineer"];
print(df);
```

```python
print(df.drop("Designation", axis = 1));
print("Main Data");
print(df);
print("Permanent remove from Main Data");
print(df.drop("Designation", axis = 1, inplace = True));# use inplace = True to delete from main data
print(df);
```

```python
print(df.loc[[0, 1]]);#for selecting rows
```

```python
print(df.iloc[3]);#for selection rows
```

```python
# for selection subsets from table
print(df.loc[[0, 1]][["Name", "City"]]);
```

```python
print(df[df["Age"]> 20]);#Age greater than 20
```

```python
print(df);
print(df[(df["Age"] > 20) & (df["City"] == "Dhanbad")]);
```
