## Pandas Series

```python
# A series is a one-Dimensional labeled array capable of holding any data type. 
# The axis labels are collectively called the index.
```

```python
import numpy as np
import pandas as pd
```

```python
labels = ["a", "b", "c", "d"];
my_list = ["Apple", "Banana", "Mango", "Coconut"];
arr = np.array([20, 60, 50, 80]);
dict = {"a":"English", "b":"Mathematics", "c":"SCience","d":"Geography"};
```

```python
print(arr);
print(dict);
print(labels);
print(my_list);
```

```python
print(pd.Series(arr, index=labels));
```

```python
print(pd.Series(dict));
```

```python
print(pd.Series(labels, index = labels));
```
