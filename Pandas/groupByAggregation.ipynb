### Group_By_Aggregation

```python
import numpy as np;
import pandas as pd;
```

```python
data = {
    "Category": ["A", "B", "A", "B", "A", "B", "A", "B"],
    "Store": ["S1", "S1", "S2", "S2","S1", "S2","S2", "S1"],
    "Sales": [100, 200, 150, 250, 120, 180, 200, 300],
    "Quantity":[10, 15, 25, 36, 18, 96, 45, 56],
    "Date": pd.date_range("2026-01-01", periods = 8)
}
df = pd.DataFrame(data);
```

```python
print(df);
```

```python
## Group by category and calculate the sum of sales
cat = df.groupby("Category");
for i, v in cat:
    print(i);
    print("\n");
    print(v);
```

```python
cat = df.groupby("Category")["Sales"].sum();
print(cat);
```

```python
# Group by Store and calculate the sum of sales
cat = df.groupby("Store");
for i, v in cat:
    print(i);
    print(v);
```

```python
cat = df.groupby("Store")["Sales"].sum();
print(cat);
```

```python
## Group by multiple columns
## Group by Category and store

cat = df.groupby(["Category","Store"])["Sales"].sum();
print(cat);
```

```python
print(df["Sales"].mean());

```

```python
print(df["Sales"].agg(["sum", "mean", "min", "count", "std", "median"]));
```
