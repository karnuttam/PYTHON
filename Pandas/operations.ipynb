## DataFrames Basic Operations

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "A": [1, 2, 3, 4, 5],
    "B": [10, 20, 30, 40, 50],
    "C": [100, 200, 300, 400, 500]
})
print(df1);
```

```python
print(df1.shape);
print(df1.columns);
```

```python
print(df1.info());
```

```python
print(df1.describe());
```

```python
def square(x):
    return x**2;
```

```python
print(df1["B"].apply(square));
```

```python
print(df1);
```

```python
df1["B"] = df1["B"].apply(square);
print(df1);
```

```python
df1["D"] =df1["B"].apply(square);
print(df1);
```

```python
df1["D"] =df1["B"].apply(lambda x: x**2);
print(df1);
```
