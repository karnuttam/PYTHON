## Count Plot

```python
"""
What's is the difference between countplot and barplot?
1.Countplot plots the count(frequency) of the number of records by category.
2.Barplot plots a value or metric for each category (by default, barplot plots the mean of a 
variable, by category).
"""
```

```python
# import libraries
import pandas as pd;
import matplotlib.pyplot as plt;
import seaborn as sns;
df = sns.load_dataset("titanic");
df
```

```python
df["class"]
```

```python
sns.countplot(x="class", data=df);
plt.title("Count of passengers by class");
plt.show();
```

```python
sns.countplot(x="class",hue="sex", data=df);
plt.title("Passenger class distributed by gender");
plt.show();
```

## Violin Plot and Box Plot

```python
"""
A violin plot is a data visualization that shows:
.Distribution of numerical data
.Density (shape) of the data
.Median and quartiles (like boxplot)
It combines (box plot and kernal density plot);
So it shows both summary statistics AND distribution shape.

| Feature                       | Violin Plot           | Box Plot      |
| ----------------------------- | --------------------- | ------------- |
| Shows Median                  |  Yes                 |  Yes         |
| Shows Quartiles               |  Yes                 |  Yes         |
| Shows Data Distribution Shape |  Yes (Density curve) |  No          |
| Detects Multimodal Data       |  Yes                 |  No          |
| Simpler to Interpret          |  Slightly complex    |  Very simple |


"""
```

```python
df = sns.load_dataset("tips");
#print(df.head())
df
```

```python
sns.boxplot(x ="day", y = "total_bill", data = df);
plt.title("Box plot of Total_bill by Day");
plt.show();
```


##  Violin plot 

```python
sns.violinplot(x="day", y ="total_bill", data = df)
plt.title("Violin plot of Total bill by Day");
plt.show();
```

```python
plt.figure(figsize=(10, 5));
plt.subplot(1, 2,1);
sns.boxplot(x="day", y="total_bill", data =df);
plt.title("Box plot");
plt.subplot(1, 2, 2);
sns.violinplot(x="day", y ="total_bill",color="orange", data = df)
plt.title("Violin plot of Total bill by Day");

plt.tight_layout();
plt.show();
```

```python
# Compare Male Vs Female Tips Distribution
sns.boxplot(x="day", y ="tip", hue="sex", palette="viridis",data=df);
plt.title("Box plot of tips by Gender");
plt.show();
```

```python
sns.violinplot(x="day", y ="tip", hue="sex", palette="Set2",data=df);
plt.title("Violin plot of tips by Gender");
plt.show();
```

## Pair Plot

```python
"""
A pair plot in Seaborn shows pairwise relationships between numerical variables in a dataset.
It creates:
.Scatter plots between every pair of numeric variables
.Distribution plot (hist/KDE) on the diagonal
It is one of the best tools for Exploratory Data Analysis (EDA).

 Why We Use Pair Plot?

To detect relationships between variables

To identify correlation patterns

To find clusters

To detect outliers

To check linear/non-linear trends

"""
```

```python
import seaborn as sns;
import matplotlib.pyplot as plt;
df = sns.load_dataset("iris");
df
```

```python
sns.pairplot(df);
plt.show()
```

```python
sns.pairplot(df, hue="species");
plt.show()
```

```python
sns.pairplot(df, hue="species", diag_kind = "kde");
plt.show()
```

```python
sns.pairplot(df, hue="species", markers=["o", "s", "D"]);
plt.show()
```

```python
sns.pairplot(df,vars =["sepal_length", "sepal_width"],kind="kde", hue="species");
plt.show()
```

## Strip Plot

```python
"""
A strip plot in Seaborn is a categorical scatter plot that shows individual data points for a numerical variable across categories.
 Why Use Strip Plot?

Use strip plot when you want to:

See individual data values

Compare distributions between categories

Detect outliers

Visualize small to medium datasets clearly
"""

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips");
df
```

```python
sns.stripplot(x ="day", y="total_bill", data=df);
plt.title("Strip plot of Total Bill by Day");
plt.show()
```

```python
sns.stripplot(x ="day", y="total_bill", data=df,hue="sex", jitter=True);
plt.title("Total Bill Distribution by Gender");
plt.show()
```

```python
"""
| Feature                  | Strip Plot | Box Plot | Violin Plot |
| ------------------------ | ---------- | -------- | ----------- |
| Shows individual points  |  Yes      |  No     |  No        |
| Shows summary statistics |  No       |  Yes    |  Yes       |
| Shows distribution shape | Partially  |  No     |  Yes       |
| Best for small datasets  |  Yes      | Moderate | Moderate    |
"""
```

```python
sns.boxplot(x="day", y="total_bill", data =df)
sns.stripplot(x ="day", y="total_bill", data =df, color="black", alpha = 0.5, jitter =True)
plt.title("Box plot with strip plot overlay");
plt.show()
```
