## Factor Plot

```python
"""
A factor plot(catplot) is a high-level categorical plotting function that:
-> Shows relationship between categorical variables
-> Can display different plot types (bar, box, violin, strip, point, count)
-> Supports multiple subplots (facets) using row and col
"""
```

```python
import seaborn as sns;
import matplotlib.pyplot as plt;
tips = sns.load_dataset("tips");
tips.head()
```

```python
# sns.set_theme(style="whitegrid", context="talk")
sns.catplot(x="day", y="total_bill", kind="bar", data=tips);
plt.show();
```

```python
sns.catplot(x="day", y="total_bill", kind="bar", data=tips, hue="sex");
plt.show();
```

```python
sns.catplot(x="day", y="total_bill",col="time",  kind="bar", data=tips);
plt.show();
```

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips);
plt.show();
```

```python
sns.catplot(x="day", y="total_bill", kind="violin", data=tips);
plt.show();
```

```python
# Custom Theme Example (Full Styling);
import seaborn as sns;
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid", context="talk", palette="pastel")
plt.figure(figsize=(8, 5));
sns.barplot(x=["Mon", "Tues","Wednes","Thurs"], y=[120, 150, 100, 170]);
plt.title("Weekly Sales", fontsize=16, fontweight="bold");
plt.xlabel("Day");
plt.ylabel("Sales");
sns.despine()
plt.savefig("graph_1.png", dpi=300);
plt.show()
```

## FacetGrid

```python
"""
FacetGrid is a seaborn class used to create multiple subplots (small multiples) based on 
categorical variables.

 It allows you to:

Split data by rows

Split data by columns

Split data by hue (color)

Apply any plotting function to each subset

It is useful when you want to compare patterns across categories visually.
"""
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
tips.head()
```

```python
# Basic syntax
"""
g = sns.FacetGrid(data, row=None, col=None, hue=None)
g.map(plot_function, x_variable, y_variable)
"""
```

```python
g =  sns.FacetGrid(tips, col="time");
g.map(sns.scatterplot, "total_bill", "tip");
plt.show();
```

```python

sns.set_context("notebook");
g = sns.FacetGrid(tips, row="sex", col="time", )
g.map(sns.scatterplot, "total_bill", "tip")

plt.show()
```

```python
g = sns.FacetGrid(tips, col="day", hue="sex")
g.map(sns.scatterplot, "total_bill", "tip")
g.add_legend()
plt.show()
```

```python
g = sns.FacetGrid(tips, col="time")
g.map(sns.histplot, "total_bill")
plt.show()
```

```python
g = sns.FacetGrid(tips, col="time",hue="sex")
g.map(sns.histplot, "total_bill")
g.add_legend();
plt.show()
```
