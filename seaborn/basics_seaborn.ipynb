## Basics of seaborn

```python
"""
Seaborn is a high-level Python Data visualization library built on top of Matplotlib that
is used to create attractive and informative statistical graphics easily.
"""
```

```python
import matplotlib.pyplot as plt;
import seaborn as sns;
import pandas as pd;
```

```python
var_1 =[1,2,3, 4, 5,6];
var_2 = [2, 4, 6, 3,2, 8];
plt.plot(var_1, var_2, marker="o");
plt.xlabel("X - axis");
plt.ylabel("y - axis");

```

```python
df = pd.DataFrame({
    "var_1":var_1,
    "var_2":var_2
});
sns.lineplot(x = "var_1", y= "var_2", data = df, marker="o");
plt.show();

```

```python
# import dataset
data = sns.load_dataset("penguins");
data
```

```python

sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data = data);

```

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data = data, hue="sex",style="sex", palette="Accent",
            markers=['o',">"]);
```

## Histogram

```python
sns.displot(data["flipper_length_mm"]);
plt.show();
```

```python
sns.displot(data["flipper_length_mm"], bins=[170, 180, 190, 200, 210, 220, 230, 240,250], rug=True, kde=True);
#kde = kernal density estimate
```

## Bar chart

```python
data_1 = sns.load_dataset("penguins");
data
```

```python
sns.barplot(x = data_1["island"], y = data_1["bill_length_mm"], data=data_1, hue="sex");
plt.show();
```

```python
sns.set(style="darkgrid");
sns.barplot(x = data_1.island, y = data_1.bill_length_mm, data=data_1,hue="sex", order=["Dream", "Biscoe", "Torgersen"]
           , hue_order=["Female", "Male"], orient="v", saturation=100, err_kws={"color":"r"});
```
