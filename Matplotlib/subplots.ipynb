## Subplots

```python
import matplotlib.pyplot as plt
```

```python
#subplot(nth row,nth column, index)
product = ["A","B", "C","D","E"];
sales = [100, 90, 95,50, 102];

plt.subplot(1,2,1);
plt.xlabel("Products");
plt.ylabel("Sales");
plt.title("Line Chart");
plt.plot(product, sales, marker="o", color="green");
plt.subplot(1, 2, 2);
plt.xlabel("Products");
plt.ylabel("Sales");
plt.title("Bar Chart");
plt.bar(product, sales, color="green");

#plt.subplots_adjust(wspace=0.5);
plt.tight_layout();
#plt.suptitle("Comparison between line and bar chart");
plt.show();
```

```python
#fig, ax = plt.subplots(nth row, nth col, figsize=(width, height))
fig, ax = plt.subplots(1,2, figsize=(10, 5));
x = [1, 2,3, 4];
y = [10, 20, 30, 40];

ax[0].plot(x,y);
ax[0].set_title("Line Plot");

ax[1].bar(x,y);
ax[1].set_title("Bar Chart");
fig.suptitle("Comparison of Line and Bar chart");
plt.tight_layout();
plt.savefig("Fig_1.png", dpi=300, bbox_inches="tight");
plt.show();
```
